### 1-(b) Each instance has three labels: Families, Genus, and Species. Each of the labels has multiple classes. We wish to solve a multi-class and multi-label problem. One of the most important approaches to multi-label classification is to train a classifier for each label (binary relevance). We first try this approach:

### 1-(b)-(i) Research exact match and hamming score/ loss methods for evaluating multilabel classification and use them in evaluating the classifiers in this problem.

**Exact Match Score:**

One trivial way around would just to ignore partially correct (consider them incorrect) and extend the accuracy used in single label case for multi-label prediction.

Exact Match Score  = 1 - Exact Match Loss

**Hamming Loss:**

It reports how many times on average, the relevance of an example to a class label is incorrectly predicted. Therefore, hamming loss takes into account the prediction error (an incorrect label is predicted) and missing error (a relevant label not predicted), normalized over total number of classes and total number of examples.

Hamming Score = 1 - Hamming Loss

### 1-(b)-(ii) Train a SVM for each of the labels, using Gaussian kernels and one versus all classifiers. Determine the weight of the SVM penalty and the width of the Gaussian Kernel using 10 fold cross validation. You are welcome to try to solve the problem with both standardized and raw attributes and report the results.

```python
#Construct X_train and X_test
X_train = df_train.iloc[:, :-4]
X_test = df_test.iloc[:, :-4]

#Construct family_train, genus_train, species_train as the labels
family_train = df_train['Family']
genus_train = df_train['Genus']
species_train = df_train['Species']

#Construct X_test, family_test, genus_test, species_test as the labels
family_test = df_test['Family']
genus_test = df_test['Genus']
species_test = df_test['Species']
```

```python
# import the SVM libraries 
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import hamming_loss, accuracy_score

# fit a SVC model with GridsearchCV to get the best C and Gamma
C_range = np.logspace(-1, 4, 5)
gamma_range = np.linspace(0.1, 2, 5)
params = {'estimator__gamma':gamma_range, 'estimator__C':C_range}
SVM = OneVsRestClassifier(SVC(kernel = 'rbf', tol = 0.7,  random_state = 0 )) 
grid = GridSearchCV(SVM, param_grid = params, cv = StratifiedKFold(10, random_state = 0, shuffle = True))
```

**SVM for Family**

```python
# Family using GridsearchCV to get the best C, Gamma, and Width of Gaussian Kernel
family_SVM = grid.fit(X_train, family_train)

#Best SVM Penalty
family_best_C = family_SVM.best_params_['estimator__C']
print('Best SVM Penalty "C":', family_best_C)

#Best SVM Gamma
family_best_Gamma = family_SVM.best_params_['estimator__gamma']
print('Best SVM Gamma:', family_best_Gamma)
```

```python
#Fit for the test set
family_pred = family_SVM.predict(X_test)

#Exact Match Score
family_score = round(accuracy_score(family_pred, family_test),3)
print('Exact Match Score:', family_score)

#Exact Match Loss
family_loss = round(1 - family_score, 3)
print('Exact Match Loss:', family_loss)

#Hamming Loss
family_Hamming_loss = round(hamming_loss(family_pred, family_test),3)
print('Hamming Loss:', family_Hamming_loss)
```

**SVM for Genus**

```python
# Genus using GridsearchCV to get the best C, Gamma, and Width of Gaussian Kernel
genus_SVM = grid.fit(X_train, genus_train)

#Best SVM Penalty
genus_best_C = genus_SVM.best_params_['estimator__C']
print('Best SVM Penalty "C":', genus_best_C)

#Best SVM Gamma
genus_best_Gamma = genus_SVM.best_params_['estimator__gamma']
print('Best SVM Gamma:', genus_best_Gamma)
```

```python
#Fit for the test set
genus_pred = genus_SVM.predict(X_test)

#Exact Match Score
genus_score = round(accuracy_score(genus_pred, genus_test),3)
print('Exact Match Score:', genus_score)

#Exact Match Loss
genus_loss = round(1 - genus_score, 3)
print('Exact Match Loss:', genus_loss)

#Hamming Loss
genus_Hamming_loss = round(hamming_loss(genus_pred, genus_test),3)
print('Hamming Loss:', genus_Hamming_loss)
```

## 2. K-Means Clustering on a Multi-Class and Multi-Label Data Set Monte-Carlo Simulation: Perform the following procedures 50 times, and report the average and standard deviation of the 50 Hamming Distances that you calculate.

### 2-(a) Use k-means clustering on the whole Anuran Calls (MFCCs) Data Set (do not split the data into train and test, as we are not performing supervised learning in this exercise). Choose k = {1; 2; : : : ; 50} automatically based on one of the methods provided in the slides (CH or Gap Statistics or scree plots or Silhouettes) or any other method you know.

```python
# read dataframe again as df1
df1 = pd.read_csv('../data/Frogs_MFCCs.csv')
df1
```

```python
# separate X and y 
X = df1.iloc[:, :-4]
y = df1.iloc[:, -4:-1]
```

**Using Silhouettes to decide the best k**

```python
import operator
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

MonteCarlo_optimal_K = {}
for t in range(1,51):
    score_dict = {}
    for k in range(2,51):
        k_means = KMeans(n_clusters = k, tol = 0.1).fit(X)
        cluster_index = k_means.labels_
        silhouettes = silhouette_score(X, cluster_index)
        score_dict[k] = silhouettes
    
    sort_silhouettes = sorted(score_dict.items(), key = operator.itemgetter(1), reverse = True)
    optimal_k = sort_silhouettes[0][0]
    MonteCarlo_optimal_K[t] = optimal_k 
```

```python
# convert to a list
k = []
for t in range(1,51):
    k.append(t)

optimalK_list = []
for t in k:
    optimalK_list.append(MonteCarlo_optimal_K[t])

Montecarlo_best_k = pd.DataFrame(np.column_stack([k, optimalK_list]))
Montecarlo_best_k = Montecarlo_best_k.rename(columns={0: 'Monte-Carlo Simulation Times', 1: 'Optimal k'})
Montecarlo_best_k
```

**According to the results, optimal k is 4.**

### 2-(b) In each cluster, determine which family is the majority by reading the true labels. Repeat for genus and species.

```python
MonteCarlo_kIndex = {}
for t in range(1, 51):
    optimal_k = MonteCarlo_optimal_K[t]
    cluster = KMeans(n_clusters = optimal_k , tol = 0.1).fit(X)    # build kmeans cluster with optimal k
    MonteCarlo_kIndex[t] = cluster.labels_
```

```python
family_majority  = []
genus_majority   = []
species_majority = []
cluster_label = []
    
for t in range(1, 51):
    optimal_k = MonteCarlo_optimal_K[t]  #optimal k in each time
    cluster_index = MonteCarlo_kIndex[t]
    
    # Majority in different clusters
    for i in range(optimal_k):
        family = {}
        genus = {}
        species = {}
        
        for j in range(len(y)):
            if cluster_index[j]==i:
                
                # Family Majority 
                family_class = y['Family'][j]
                if family_class in family.keys():
                    family[family_class] += 1
                else:
                    family[family_class] = 1
                
                # Genus Majority
                genus_class = y['Genus'][j]
                if genus_class in genus.keys():
                    genus[genus_class] += 1
                else:
                    genus[genus_class] = 1
                
                # Species Majority
                species_class = y['Species'][j]
                if species_class in species.keys():
                    species[species_class] += 1
                else:
                    species[species_class] = 1
        
        #Sort the label
        sorted_family = sorted(family.items(), key = operator.itemgetter(1), reverse = True)
        sorted_genus = sorted(genus.items(), key = operator.itemgetter(1), reverse = True)
        sorted_species = sorted(species.items(), key = operator.itemgetter(1), reverse = True)
        
        label = {}
        label['Family'] = sorted_family[0][0]
        label['Genus'] = sorted_genus[0][0]
        label['Species'] = sorted_species[0][0]
        cluster_label.append(label)
        
        #Find the majority
        family_majority.append(sorted_family[0][0])
        genus_majority.append(sorted_genus[0][0])
        species_majority.append(sorted_species[0][0]) 
```

```python
#Times and Clusters
times = []
cluster_list = []

for t in range(1, 51):
    for i in range(MonteCarlo_optimal_K[t]):
        times.append(t)
        cluster_list.append('Cluster ' + str(i+1))

#Family Majority, Genus Majority, Species Majority
Get_Majority_table = pd.DataFrame(np.column_stack([times, cluster_list, family_majority, genus_majority, species_majority]))
Get_Majority_table = Get_Majority_table.rename(columns={0: 'Monte-Carlo Simulation Times', 1: 'Cluster',
                                                   2: 'Family Majority', 3: 'Genus Majority', 4:'Species Majority' })
Get_Majority_table
```

### 2-(c)  Now for each cluster you have a majority label triplet (family, genus, species). Calculate the average Hamming distance, Hamming score, and Hamming loss between the true labels and the labels assigned by clusters.

**For the Hamming Distance, I use the sum of the Hamming loss in family, genus, and species; then using average of Hamming Distance as the Hamming_loss.**

**Hamming Distance:**

Hamming Distance  = family_Hamming_loss + genus_Hamming_loss + species_Hamming_loss

**Hamming Loss:**

Hamming Loss = Hamming Distance/3

**Hamming Score:**

Hamming Score = 1 - Hamming Loss

```python
Hamming_Distance_list = []
Hamming_Loss_list  = []
Hamming_Score_list = []

for t in range(1, 51):
    cluster_index = MonteCarlo_kIndex[t]
    family_pred  = []
    genus_pred   = []
    species_pred = []
    
    for i in range(len(X)):
        cluster_name = cluster_index[i]
        label = cluster_label[cluster_name]
        family_pred.append(label['Family'])
        genus_pred.append(label['Genus'])
        species_pred.append(label['Species'])
    
    #Family, Genus and Species
    family = np.array(y['Family'])
    genus = np.array(y['Genus'])
    species = np.array(y['Species'])
    
    #Prediction of Family, Genus and Species, y
    family_pred = np.array(family_pred)
    genus_pred = np.array(genus_pred)
    species_pred = np.array(species_pred)
    
    #Family Hamming Loss and Family Hamming Score
    family_Hamming_loss = hamming_loss(family_pred, family)/3
    family_hamming_score = 1 - family_Hamming_loss
    
    #Genus Hamming Loss and Genus Hamming Score
    genus_Hamming_loss = hamming_loss(genus_pred, genus)/3
    genus_hamming_score = 1 - genus_Hamming_loss
    
    #Species Hamming Loss and Species Hamming Score
    species_Hamming_loss = hamming_loss(species_pred, species)/3
    species_hamming_score = 1 - species_Hamming_loss
    
    #Hamming Distance, Hamming Loss, Hamming Score
    Hamming_Distance = family_Hamming_loss + genus_Hamming_loss + species_Hamming_loss
    Hamming_Distance = round(Hamming_Distance, 3)
    
    Hamming_Loss = Hamming_Distance/3
    Hamming_Loss = round(Hamming_Loss, 3)
    
    Hamming_Score = 1- Hamming_Loss
   # Hamming_Score = (family_hamming_score + genus_hamming_score + species_hamming_score)/3
    Hamming_Score = round(Hamming_Score, 3)
    
    Hamming_Distance_list.append(Hamming_Distance)
    Hamming_Loss_list.append(Hamming_Loss)
    Hamming_Score_list.append(Hamming_Score)
```

```python
#Hamming Distance, Hamming Loss, Hamming Score
T = []
for t in range(1, 51):
    T.append(t)
    
#np.column_stack
    
Monte_Carlo_table =pd.DataFrame(np.column_stack([T, Hamming_Distance_list, Hamming_Loss_list, Hamming_Score_list]))
Monte_Carlo_table = Monte_Carlo_table.rename(columns={0: 'Monte-Carlo Simulation Times', 1: 'Hamming Distance',
                                                     2: 'Hamming Loss', 3: 'Hamming Score'})
Monte_Carlo_table
```

```python
table = pd.DataFrame(np.zeros((2,3)))
table.columns = ['Hamming Distance', 'Hamming Loss', 'Hamming Score']
table.index = ['Average', 'Std']

mean_Hamming_d = round(np.mean(Hamming_Distance_list),3)
table.iloc[0,0] = float(mean_Hamming_d)

std_Hamming_d = round(np.std(Hamming_Distance_list),3)
table.iloc[1,0] = float(std_Hamming_d)

mean_Hamming_l = round(np.mean(Hamming_Loss_list),3)
table.iloc[0,1]  = float(mean_Hamming_l)

std_Hamming_l = round(np.std(Hamming_Loss_list),3)
table.iloc[1,1] = float(std_Hamming_l)

mean_Hamming_s = round(np.mean(Hamming_Score_list),3)
table.iloc[0,2] = float(mean_Hamming_s)

std_Hamming_s = round(np.std(Hamming_Score_list),3)
table.iloc[1,2] = float(std_Hamming_s)
table
```
